# For SEX Column

```python
insurance_datset["sex"] = insurance_datset["sex"].apply(lambda x:0 if x=="female" else 1) 
insurance_datset.head()
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
insurance_datset["sex"] = label_encoder.fit_transform(insurance_datset["sex"])
print(insurance_datset.head())
```

# For Smoker Column

```python
# insurance_datset["smoker"] = insurance_datset["smoker"].apply(lambda x:0 if x == "no" else 1)
```

```python
insurance_datset["smoker"] = label_encoder.fit_transform(insurance_datset["smoker"])
print(insurance_datset.head())
```

# For region column 

```python
insurance_datset["region"].unique()
```

```python
region_dumies = pd.get_dummies(insurance_datset["region"])
region_dumies.head()
```

```python
insurance_datset = pd.concat([insurance_datset,region_dumies],axis=1)
insurance_datset.drop("region",axis=1,inplace = True)
insurance_datset.head()
```

# Training and Spliting Our Data

```python
from sklearn.model_selection import train_test_split 
from sklearn.preprocessing import StandardScaler,MinMaxScaler

X = insurance_datset.drop("charges",axis=1).values
y= insurance_datset.charges.values.reshape(-1,1)

x_label = StandardScaler()
X = x_label.fit_transform(X)


y_label = StandardScaler()
y = y_label.fit_transform(y)


X_train,X_test , y_train,y_test = train_test_split(X,y,test_size=.2,random_state=42) 
print(f"The Length of testing data =  {len(X_test)}")
print(f"The Length of training data =  {len(X_train)}")


```

```python
from sklearn.linear_model import LinearRegression

linear_regression = LinearRegression()
linear_regression.fit(X_train,y_train)


# check the accuracy of linear regression
print("Score : ",end="")
print(linear_regression.score(X_test,y_test))
```
