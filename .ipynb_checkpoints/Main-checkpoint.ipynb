# Step 3 Data Preprocessing

```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import warnings
from sklearn.preprocessing import LabelEncoder, StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns
warnings.filterwarnings("ignore")

df = pd.read_csv("insurance.csv")
print(df.info())
# print(df.describe(include="all"))
# print(df.shape)
print(df.head())
print(df.columns)
```

```python
le = LabelEncoder()
df['sex']=le.fit_transform(df['sex'])
df['smoker']=le.fit_transform(df['smoker'])
df['region']=le.fit_transform(df['region'])
print(df.head())
print(df.info())
```

# Checking the unique values

```python
dict={}
for i in list(df.columns):
    dict[i]=df[i].value_counts().shape[0]
# print(dict)
pd.DataFrame(dict,index=["Unique counts"]).transpose()
```

# Step 3 Exploratory Data Analysis (EDA)

```python
import matplotlib.pyplot as plt
import seaborn as sns

fig = plt.figure(figsize=(15,8))
gs = fig.add_gridspec(2,2)
gs.update(wspace=0.5, hspace=1.25)

x1 = fig.add_subplot(gs[0,0])
x2 = fig.add_subplot(gs[0,1])
x3 = fig.add_subplot(gs[1,0])
x4 = fig.add_subplot(gs[1,1])

sns.barplot(ax=x1,data=df,x='charges',y = 'age')
sns.barplot(ax=x2,data=df,x='sex',y = 'charges')
sns.barplot(ax=x3,data=df,x='charges',y = 'bmi')
sns.barplot(ax=x4,data=df,x='children',y = 'charges')
plt.show()
```

```python
from patsy import dmatrices
from statsmodels.stats.outliers_influence import variance_inflation_factor

y, X = dmatrices('charges ~ age+bmi+sex+children+smoker+region+charges', data=df, return_type='dataframe')

vif = pd.DataFrame()
vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif['variable'] = X.columns


vif
```
